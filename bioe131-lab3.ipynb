```python
import matplotlib.pyplot as plt

dictionary = {}

for i in open('romeoandjuliet.txt'):
    for j in i.split():
        k = j.lower().replace(',','').replace('.','')
        if k in dictionary:
            dictionary[k] = dictionary[k] + 1
        else:
            dictionary[k] = 1
```

```python
plt.bar(range(10), [x[1] for x in sorted(dictionary.items(), key=lambda x: -x[1])[0:10]], align='center')
plt.xticks(range(10), [x[0] for x in sorted(dictionary.items(), key=lambda x: -x[1])[0:10]])
plt.show()
```
